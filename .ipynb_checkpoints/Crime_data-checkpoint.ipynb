```python
import numpy as np
import pandas as pd
```

```python
# import the dataset
df = pd.read_csv('crime_data.csv')
```

```python
# check the fist 5 rows
df.head()
```

```python
# check the last 5 rows
df.tail()
```

```python
# specify the datatype for shooting
# drop UCR_PART
df.drop(columns='UCR_PART')
df.dropna
```

```python
"""
    1.What crime is more popular in Boston? 
    ie. What offense is more popular
    get the count
""" 
# get the 5 top popular crimes
df.groupby(['OFFENSE_CODE'])['OFFENSE_CODE_GROUP'].value_counts().nlargest(5)
```

```python
crime_popularity =df.groupby(['OFFENSE_CODE'])['OFFENSE_CODE_GROUP'].value_counts().nlargest(15).sort_values()
crime_popularity.plot.bar()
```

```python
# What hour of the day, and which day of the week is more prone to crime? 
# Day
mostPopularCrimeDay = df['DAY_OF_WEEK'].value_counts().nlargest(7).sort_values()
mostPopularCrimeDay.plot.bar()
```

```python
mostPopularCrimeDay = df['DAY_OF_WEEK'].mode()
mostPopularCrimeDay
```

```python
hour_by_crimes = df['HOUR'].value_counts().nlargest(24).sort_values()
hour_by_crimes.plot.bar()
```

```python
mostPopularCrimeHour = df['HOUR'].mode()
mostPopularCrimeHour
```
