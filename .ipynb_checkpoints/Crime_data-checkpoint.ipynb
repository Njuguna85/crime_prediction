```python
"""
    1.What crime is more popular in Boston? 
    ie. What offense is more popular
    get the count
""" 
# get the 5 top popular crimes
df.groupby(['OFFENSE_CODE'])['OFFENSE_CODE_GROUP'].value_counts().nlargest(5)
```

```python
crime_popularity =df.groupby(['OFFENSE_CODE'])['OFFENSE_CODE_GROUP'].value_counts().nlargest(15).sort_values()
crime_popularity.plot.bar()
```

```python
# What hour of the day, and which day of the week is more prone to crime? 
# Day
mostPopularCrimeDay = df['DAY_OF_WEEK'].value_counts().nlargest(7).sort_values()
mostPopularCrimeDay.plot.bar()
```
